```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#function and gradient setup
def f(x,y):
    return (x-3)**2 + 2*(y-1)**2

def gradient_f(x,y):
    df_dx=2*(x-3)
    df_dy=4*(y-1)
    return np.array([df_dx,df_dy])

gradient_f(3,1)
```

```python
#update rule

def update(alpha = 0.1, max_iterations = 1000, tolorance = 1e-6):
    pos = np.array([0.0,0.0])
    history = [pos.copy()]

    for i in range(max_iterations):
        grad = gradient_f(pos[0], pos[1])
        magnitude = np.linalg.norm(grad)
        if(magnitude < tolorance):
            print(f'Converged at interation {i} at point {pos[0]:0.4f} {pos[1]:0.4f}')
            return history
        else:
            pos = pos-(alpha*grad)
            history.append(pos.copy())

        
    
history = update()
```

```python
x_range= np.linspace(-1,6,100)
y_range = np.linspace(-2,4,100)
X,Y = np.meshgrid(x_range, y_range)
Z = f(X, Y)

plt.figure(figsize=(10,8))
plt.contour(X,Y,Z, levels=30, cmap='viridis')
plt.colorbar(label='f(x,y)')

history = np.array(history)
plt.plot(history[:,0], history[:,1], 'ro-', markersize=3)
plt.plot(history[0,0], history[0,1], 'go', markersize=10)
plt.plot(history[-1,0], history[-1,1], 'r*', markersize=15)

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Gradient Descent Path')
plt.show()

```
