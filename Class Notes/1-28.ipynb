```python
import torch
import sympy as sp
```

```python
#x is a single variable with 1 value
x = torch.tensor(2.0, requires_grad=True)
x.shape
```

```python
#Y is a function that uses X as a value
y =-1*(0.5)*x**3 + 1
y.type
```

```python
# .backward() computes the gradient for all tensors involved in another tensor
y.backward()
print(x.grad)
```

```python
x_sym = sp.symbols('x')
y_sym = -1*(0.5)*x_sym**3 + 1
print(y_sym)
```

```python
dydx = sp.diff(y_sym, x_sym)
dydx
```
